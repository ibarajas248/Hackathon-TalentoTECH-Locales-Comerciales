```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim
import time
import re
from IPython.display import display

# Cargar el archivo Excel en un DataFrame
df = pd.read_excel('mercado.xlsx')

# Filtrar el DataFrame para incluir solo las filas con dirección no nula
df = df[df['Dirección'].notna()]

# Tomar los primeros 20 registros
df_10 = df.head(2000)

# Crear un mapa centrado en Bogotá
mapa = folium.Map(location=[4.6101, -74.0817], zoom_start=12)

# Inicializar el geocodificador
geolocator = Nominatim(user_agent="mi_geocodificador")

# Función para limpiar la dirección
def limpiar_direccion(direccion):
    # Eliminar "Piso" y lo que le sigue
    direccion_limpia = re.sub(r',? Piso \d+', '', direccion)  # Elimina "Piso" seguido de un número
    return direccion_limpia

# Crear una lista para almacenar las coordenadas
coordenadas = []

# Agregar coordenadas al mapa para los primeros 20 registros
for index, row in df_10.iterrows():
    # Limpiar la dirección para quitar información de piso
    direccion = limpiar_direccion(f"{row['Dirección']}, Bogotá")
    
    try:
        # Intentar geocodificar la dirección
        location = geolocator.geocode(direccion, timeout=10)  # Timeout de 10 segundos
    except Exception as e:
        print(f"Error al geocodificar '{direccion}': {e}")
        continue  # Continuar con la siguiente dirección

    if location:
        # Almacenar la ubicación en la lista si es exitosa
        coordenadas.append([location.latitude, location.longitude])
        print(f"Georeferenciado exitoso: {direccion}")
    else:
        print(f"No se pudo geocodificar la dirección: {direccion}")

    # Pausa de 1 segundo entre solicitudes para evitar limitaciones
    time.sleep(1)

# Agregar el mapa de calor si hay coordenadas
if coordenadas:
    HeatMap(coordenadas).add_to(mapa)
    print(f"Se agregaron {len(coordenadas)} direcciones al mapa de calor.")
else:
    print("No hay coordenadas válidas para mostrar en el mapa de calor.")

# Mostrar el mapa en Jupyter Notebook
display(mapa)
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim, GoogleV3
import time
import re
import spacy  # Librería para el procesamiento de lenguaje natural
from IPython.display import display

# Cargar el archivo Excel en un DataFrame
df = pd.read_excel('mercado.xlsx')

# Filtrar el DataFrame para incluir solo las filas con dirección no nula
df = df[df['Dirección'].notna()]

# Tomar los primeros 20 registros
df_10 = df.head(1000)

# Crear un mapa centrado en Bogotá
mapa = folium.Map(location=[4.6101, -74.0817], zoom_start=12)

# Inicializar el geocodificador de Nominatim
geolocator = Nominatim(user_agent="mi_geocodificador")

# Si tienes acceso a la API de Google, puedes usarla para mejorar la geocodificación
# google_geolocator = GoogleV3(api_key="YOUR_GOOGLE_API_KEY")

# Inicializar spaCy para NLP (modelo en español)
nlp = spacy.load("es_core_news_sm")

# Función para limpiar la dirección
def limpiar_direccion(direccion):
    # Eliminar "Piso" y lo que le sigue
    direccion_limpia = re.sub(r',? Piso \d+', '', direccion)  # Elimina "Piso" seguido de un número
    return direccion_limpia

# Función para corregir direcciones con NLP usando spaCy
def corregir_direccion(direccion):
    doc = nlp(direccion)
    # Identificar entidades geográficas como direcciones o lugares en el texto
    entidades = [ent.text for ent in doc.ents if ent.label_ == "LOC"]
    if entidades:
        # Devolver la entidad geográfica identificada como corrección
        return ', '.join(entidades) + ', Bogotá'
    else:
        return direccion

# Crear una lista para almacenar las coordenadas
coordenadas = []

# Agregar coordenadas al mapa para los primeros 20 registros
for index, row in df_10.iterrows():
    # Limpiar la dirección para quitar información de piso
    direccion = limpiar_direccion(f"{row['Dirección']}, Bogotá")
    
    # Intentar geocodificar la dirección con Nominatim
    try:
        location = geolocator.geocode(direccion, timeout=10)  # Timeout de 10 segundos
    except Exception as e:
        print(f"Error al geocodificar '{direccion}': {e}")
        continue  # Continuar con la siguiente dirección

    # Si no se puede geocodificar, intentar mejorarla con NLP
    if not location:
        print(f"No se pudo geocodificar la dirección original: {direccion}. Intentando corregir...")
        direccion_corregida = corregir_direccion(direccion)
        
        try:
            # Reintentar la geocodificación con la dirección corregida
            location = geolocator.geocode(direccion_corregida, timeout=10)
            if location:
                print(f"Geocodificación exitosa tras corrección: {direccion_corregida}")
            else:
                print(f"No se pudo geocodificar tras corrección: {direccion_corregida}")
        except Exception as e:
            print(f"Error al geocodificar '{direccion_corregida}': {e}")
            continue
    
    # Almacenar la ubicación en la lista si es exitosa
    if location:
        coordenadas.append([location.latitude, location.longitude])
    
    # Pausa de 1 segundo entre solicitudes para evitar limitaciones
    time.sleep(1)

# Agregar el mapa de calor si hay coordenadas
if coordenadas:
    HeatMap(coordenadas).add_to(mapa)
    print(f"Se agregaron {len(coordenadas)} direcciones al mapa de calor.")
else:
    print("No hay coordenadas válidas para mostrar en el mapa de calor.")

# Mostrar el mapa en Jupyter Notebook
display(mapa)
```

```python
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim
import requests
import pandas as pd
from IPython.display import display

# Crear un geocodificador con un user_agent único
geolocator = Nominatim(user_agent="mi_aplicacion_geocodificacion_v1.0")

# Dirección a geocodificar
direccion = "lourdes chapinero "

# Obtener la ubicación
try:
    ubicacion = geolocator.geocode(direccion)

    if ubicacion:
        # Coordenadas de la ubicación
        latitud = ubicacion.latitude
        longitud = ubicacion.longitude

        # Crear un mapa centrado en la ubicación
        mapa = folium.Map(location=[latitud, longitud], zoom_start=14)

        # Agregar un marcador en la ubicación
        folium.Marker(
            location=[latitud, longitud],
            popup=direccion,
            icon=folium.Icon(color='blue')
        ).add_to(mapa)

        # Realizar una consulta a Overpass API para lugares de interés
        query = f"""
        [out:json];
        (
          node["amenity"](around:1000, {latitud}, {longitud});
          way["amenity"](around:1000, {latitud}, {longitud});
          relation["amenity"](around:1000, {latitud}, {longitud});
        );
        out body;
        """
        overpass_url = "http://overpass-api.de/api/interpreter"
        response = requests.get(overpass_url, params={'data': query})

        if response.status_code == 200:
            data = response.json()

            # Crear una lista para almacenar los lugares y sus datos
            lugares_interes = []
            coordenadas_heatmap = []  # Lista para coordenadas del mapa de calor

            # Agregar lugares de interés al mapa y a la lista
            for element in data['elements']:
                if 'tags' in element and 'lat' in element and 'lon' in element:
                    # Extraer información básica
                    nombre = element['tags'].get('name', 'Sin nombre')
                    amenidad = element['tags'].get('amenity', 'Desconocido')
                    direccion = element['tags'].get('address', 'Sin dirección')

                    # Agregar a la lista para el mapa de calor
                    coordenadas_heatmap.append([element['lat'], element['lon']])

                    # Agregar a la lista de lugares de interés
                    lugares_interes.append({
                        'Lugar': nombre,
                        'Tipología': amenidad,
                        'Dirección': direccion
                    })

            # Crear un DataFrame con los lugares de interés
            df_lugares = pd.DataFrame(lugares_interes)

            # Mostrar el DataFrame
            display(df_lugares)

            # Crear el mapa de calor
            HeatMap(coordenadas_heatmap).add_to(mapa)

            # Agregar leyenda de colores al mapa con funcionalidad desplegable
            legend_html = '''
            <div style="position: fixed;
                        bottom: 50px; left: 50px; width: 200px; height: auto;
                        border:2px solid grey; z-index:9999; font-size:14px;
                        background-color:white; padding: 10px; display: none;" id="legend">
                <b>Convenciones de Colores</b><br>
                <i class="fa fa-circle" style="color: orange"></i>&nbsp; Restaurant<br>
                <i class="fa fa-circle" style="color: green"></i>&nbsp; Café<br>
                <i class="fa fa-circle" style="color: blue"></i>&nbsp; Escuela<br>
                <i class="fa fa-circle" style="color: red"></i>&nbsp; Hospital<br>
                <i class="fa fa-circle" style="color: purple"></i>&nbsp; Bar<br>
                <i class="fa fa-circle" style="color: lightblue"></i>&nbsp; Banco<br>
                <i class="fa fa-circle" style="color: darkgreen"></i>&nbsp; Farmacia<br>
                <i class="fa fa-circle" style="color: darkred"></i>&nbsp; Mercado<br>
            </div>
            <button onclick="var legend = document.getElementById('legend'); legend.style.display = (legend.style.display === 'none') ? 'block' : 'none';"
                    style="position: fixed; bottom: 10px; left: 50px; z-index:9999;">
                Leyenda
            </button>
            '''
            mapa.get_root().html.add_child(folium.Element(legend_html))

            # Mostrar el mapa directamente en Colab
            display(mapa)

        else:
            print("Error al consultar la API de Overpass:", response.status_code)

    else:
        print("La dirección no se pudo geocodificar.")
except Exception as e:
    print("Ocurrió un error:", str(e))
```
